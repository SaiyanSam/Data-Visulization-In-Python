```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\HP\\Downloads\\Concrete_Data.csv')

## syntax for line graph
## figsize can be used in all plots and it can change the size of the graph
df.drop(['Cement'], axis=1).plot.line(title='Line graph example',figsize=(15,5))
plt.ylabel('Frequency')
plt.xlabel('Values')
plt.show()
```
