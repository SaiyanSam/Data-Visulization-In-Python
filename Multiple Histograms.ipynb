```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:\\Users\\HP\\Downloads\\Concrete_Data.csv')

## To generate histogram for each column
## Here we have to give layout argument depending of columns ex: 3X3=9 columns
## figsize define size of each histogram
## bins is related to width of each histogram bar section. More the bins : more blocks there will be

df.plot.hist(subplots=True, layout=(3,3), figsize=(10, 10), bins=30)
##To show the plot created. xlabel and ylabel can be changed
plt.show()
```
