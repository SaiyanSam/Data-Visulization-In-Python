```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

data = pd.read_csv("F://Reading material//Data Science using Python//Youtub source//Linear regression using single variable//Data//Prices.csv")
##Generation of regression model
reg = linear_model.LinearRegression()
reg.fit(data[['price']],data.area)

##To create a scatter plot
%matplotlib inline
plt.xlabel('Price (in Rupee)')
plt.ylabel('Area (in sq. m)')
plt.scatter(data[['price']],data.area)

##To show the best fit line :
plt.plot(data.price,reg.predict(data[['price']]),color='red')
```
